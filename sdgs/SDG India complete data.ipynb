```python
indicators = pd.read_excel(open('F:/Pykih/SDG Data/Goals, Target, and Indicators.xlsx','rb'),sheet_name='Targets')
```

```python
indicators.head(50)
```

```python
number = list(indicators['Target'])
target_lists = []
i = 0
for num in number:
    strin_number = str(num).strip()
    strin_number = strin_number.split(".")
    strin_number = strin_number[0]
    target_lists.append(int(strin_number))

a = target_lists
indicators['SDG'] = a
```

```python
indicators.head()
```

```python
indicators.shape 
```

```python
final_data= data3.join(indicators.set_index('SDG'), on='SDG')
```

```python
final_data.head(50)
```

```python
final_data.shape
```

```python
columnsTitles = ['SDG', 'SDG GoalShort', 'SDG GoalLong','Target','Description','Indicator','States_Uts','Value']

final_data = final_data.reindex(columns=columnsTitles)
```

```python
final_data.head()
```

```python
# convert csv to xlsx file
writer = pd.ExcelWriter('SDG India Index Complete data.xlsx')
final_data.to_excel(writer, index = False)
writer.save()
```
